# Answer

```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
print("Setup Complete")
```

```python
california_wildfire_filepath = "Updated_California_Wildfire_Damage.csv"

california_wildfire = pd.read_csv(california_wildfire_filepath , index_col = 'Incident_ID')
```

```python
california_wildfire.head()
```

```python
# histogram

# plt.figure(figsize = (5,5))

sns.histplot(y = california_wildfire['Location'])

plt.title("Total Incidents in a county")
```

Analysis:

Wildfires happen mostly in Shasta and Sonoma counties, while Riverside, Santa Barbara, and LA counties face fewer wildfires.

```python
# pairplot

sns.pairplot(california_wildfire , hue = 'Location')

```

Analysis:

Most areas are burnt in Sonoma County.

Most homes are destroyed in Shasta County.

Most businesses are destroyed in Butte County.

Most vehicles are destroyed in Shasta County.

Every county has a large amount of injuries.

Most fatalities occurred in San Diego County.

Blutte County faces the highest economic loss as most businesses get destroyed in here.

```python
# Heatmap

california_wildfire_customized = california_wildfire.drop(['Area_Burned_Acres' , 'Homes_Destroyed' , 'Estimated_Financial_Loss_Million_' , 'Date' ,'Location','Cause'] , axis = 1)
```

```python
california_wildfire_customized.head()
```

```python
#heatmap
plt.figure(figsize = (10,20))
sns.heatmap(data = california_wildfire_customized , annot = True)
```

Analysis:

While the number of fatalities and injuries is low, business and vehicle losses are huge.
Local government and central government should discover a safe exit for vehicles and discover a way to lessen the economic loss.

```python
#jointplot 

sns.jointplot(x = 'Estimated_Financial_Loss_Million_' , y = 'Area_Burned_Acres' , data = california_wildfire , kind = 'kde')
```

Analysis:

Many times the amount of burnt areas stays at 10-20 thousand acres, having an estimated financial loss of 1-2 thousand million dollars. But this number can go as high as 25-32 thousand acres, having an estimated financial loss of 3500 million dollars.

```python
#Bar plot

california_wildfire_limited = california_wildfire.head(25)

sns.barplot(x = california_wildfire_limited.index , y = california_wildfire_limited['Fatalities'])
```

Analysis:

```python
#violin plot

sns.violinplot(y = california_wildfire['Estimated_Financial_Loss_Million_'])
```

Analysis:

On average, every incident costs 2500 million dollars. Most of the incident financial losses in range of 3500 to 4000 million dollars and 1500 to 2000 million dollars.
