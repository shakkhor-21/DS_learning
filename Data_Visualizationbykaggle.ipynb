# Lecture 1: Hello Seaborn

1. Setting up the workplace

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
print("Setup Completed!")
```

2. Importing the dataset

```python
filepath = "fifa.csv"
fifa_data = pd.read_csv(filepath, index_col = "Date" , parse_dates = True )
```

```python
#examining the dataset
fifa_data.head(17)
```

```python
#customizing figure size
plt.figure(figsize=(16,6))

#lineplot using seaborn
sns.lineplot(data = fifa_data)
```

# Lecture 02 : Line Charts

```python
sp_filepath = "spotify.csv"
spotify = pd.read_csv(sp_filepath , index_col = "Date" , parse_dates = True)
print(spotify.head())
# print('********************************************************************')
spotify.tail()
```

```python
#line plot by using seaborn

sns.lineplot(data = spotify)
```

```python
plt.figure(figsize=(14,6))
plt.title("Spotify Top Songs Stats")

sns.lineplot(data = spotify)
```

```python
# name of the columns
list(spotify.columns)
```

```python
plt.figure(figsize = (14,6))

plt.title("Spotify Top Songs")

sns.lineplot(data = spotify['Shape of You'] , label = 'Shape of You')

sns.lineplot(data = spotify['Despacito'] , label = 'Despacito')

plt.xlabel("Date")
```

# Lecture 3

```python
file_path = "flight_delays.csv"

flight_delays = pd.read_csv(file_path , index_col = 'Month')
```

```python
flight_delays
```

```python
plt.figure(figsize = (10,6))

plt.title("Average Flight Delay")

sns.barplot(x = flight_delays.index , y = flight_delays['NK'])

plt.ylabel("Delay in Minutes")
```

# Heatmap

```python
plt.figure(figsize = (10,6))

plt.title("Average Delay by month Heatmap")

sns.heatmap(data = flight_delays , annot = True)

plt.xlabel("Ailine")
```

# Lecture 4 : Scatterplot

```python
filepath = "insurance.csv"

insurance = pd.read_csv(filepath)
```

```python
insurance.head()
```

```python
sns.scatterplot(x = insurance['age'] , y = insurance['bmi'])
```

```python
sns.regplot(x = insurance['age'] , y = insurance['bmi'])
```

```python
sns.scatterplot(x = insurance['bmi'] , y = insurance['charges'] , hue = insurance['smoker'])
```

```python
sns.lmplot(x = 'bmi' , y = 'charges', hue = 'smoker' , data = insurance)
```

```python
sns.swarmplot(y = insurance['smoker'] , x = insurance['charges'])
```

# Lecture 5: Distributions

# Histogram

```python
iris_filepath = "iris.csv"

iris = pd.read_csv(iris_filepath , index_col = 'Id')
```

```python
iris.head()
```

```python
sns.histplot(iris['Petal Length (cm)'])
```

# Density Plots

```python
sns.kdeplot(data = iris['Petal Length (cm)'] , shade = True)
```

```python
sns.jointplot(x = iris['Petal Length (cm)'] , y = iris['Sepal Width (cm)'] , kind = 'kde')
```

# Color Coded Plots

```python
sns.histplot(data = iris , x = iris['Petal Length (cm)'] , hue = iris['Species'])
```

```python
sns.kdeplot(x = iris['Petal Length (cm)'] , hue = iris['Species'] , shade = True )
```
