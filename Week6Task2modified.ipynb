Question: Briefly describe the dataset in your notebook, including the context, number of rows and columns, and a short explanation of the columns.

Answer: 

This dataset provides annual statistics on U.S. immigration from fiscal years 1980 to 2021. All data are official figures sourced from the U.S. Department of Homeland Security's government website, compiled and structured by me. The decision to focus on data from 1980 onwards is based on several factors. Starting in 1976, the U.S. government fiscal year began on October 1 and ended on September 30 of the following year. Including data from before 1976 could lead to inaccuracies and confusion. Additionally, the U.S. only began tracking refugee arrivals after the Refugee Act of 1980, a key statistic included in this dataset. For these reasons, the dataset begins in 1980 rather than 1976 and includes the following attributes:

Immigrants Obtaining Lawful Permanent Resident Status - Number of immigrants who obtained lawful permanent resident status in the United States, otherwise known as green card holders.

Refugee Arrivals - Number of refugees who arrived in the United States. Excludes Amerasian immigrants except for the fiscal years of 1989 and 1991. Figures are based on refugee's arrival date.

Noncitizen Apprehensions - Number of noncitizens apprehended in the United States. Data from 2020 to 2021 includes U.S. Customs and Border Protection (CBP) encounters that resulted in expulsion on public health grounds (due to the pandemic).

Noncitizen Removals - Number of noncitizens removed from the United States. Removals are the compulsory and confirmed movement of an inadmissible or deportable noncitizen out of the United States based on an order of removal.

Noncitizen Returns - Number of noncitizen returns from the United States. Returns are the confirmed movement of an inadmissible or deportable noncitizen out of the United States not based on an order of removal.

# 2. Basic Plots (40 Marks)

Line Plot: Show a trend or progression of a variable over time or another index. (10 Marks)

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
Immigration = pd.read_csv("Immigration_Clean.csv")
```

```python
print(Immigration.head())
```

```python
plt.figure(figsize=(10, 5))
plt.plot(Immigration['Year'],Immigration['Immigrants Obtaining Lawful Permanent Resident Status'])
plt.xlabel("Year")
plt.ylabel("Obtaining Lawful Permanent Resident Status")
plt.title("Immigration Trends over time")
plt.show()
```

Bar Chart: Compare categories or groups. (10 Marks)

```python
plt.figure(figsize=(10,5))
plt.bar(Immigration['Year'],Immigration['Refugee Arrivals'])
plt.show()
```

Scatter Plot: Illustrate relationships between two numeric variables. (10 Marks)

```python
plt.scatter(Immigration['Year'],Immigration['Noncitizen Apprehensions'],marker = 'o')
plt.scatter(Immigration['Year'],Immigration['Noncitizen Removals'] , marker = '4')
plt.legend(['Noncitizen Apprehensions','Noncitizen Removals'] , loc = 'best')
plt.show()
```

Box Plot: Summarize the distribution and identify outliers of a numeric variable. (10 Marks)

```python
Immigration['Noncitizen Returns'].describe()
```

```python
plt.boxplot(Immigration['Noncitizen Returns'])
plt.show()
```

# 3. Customizing Plots (20 Marks)

Add meaningful titles, labels, and legends to all the plots. (10 Marks)
Experiment with colors and markers to make your plots more readable and visually appealing. (10 Marks)

```python
#lineplot
plt.plot(Immigration['Year'],Immigration['Immigrants Obtaining Lawful Permanent Resident Status'] , color = 'r')
plt.xlabel("Year")
plt.ylabel("Obtaining Lawful Permanent Resident Status")
plt.title("Immigration Trends over time")
plt.legend(['Obtaining Lawful Permanent Resident Status'] , loc = 'best')
plt.show()
```

```python
#boxplot
plt.xlabel("Year")
plt.ylabel("Refugee Arrivals")
plt.title("Immigration - Refugees")
plt.bar(Immigration['Year'],Immigration['Refugee Arrivals'] , color ='g')
plt.legend(['Refugee Arrivals'] , loc = 'best')
plt.show()
```

```python
#scatterplot
plt.xlabel("Year")
plt.title("Apprehensions & Removals over time")
plt.scatter(Immigration['Year'],Immigration['Noncitizen Apprehensions'],marker = 'o' , s = 200 , c = 'r' , edgecolor = 'b' , alpha = 0.5)
plt.scatter(Immigration['Year'],Immigration['Noncitizen Removals'] , marker = '4' , s = 150 , c = 'blue' , alpha = 0.5)
plt.legend(['Noncitizen Apprehensions' , 'Noncitizen Removals'] , loc = 'best')
plt.show()
```

```python
#boxplot
plt.title("Returns")
plt.ylabel("Non-Citizen Returns")
plt.boxplot(Immigration['Noncitizen Returns'])
plt.show()
```
