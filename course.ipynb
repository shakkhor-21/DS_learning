```python
import numpy as np
from matplotlib import pyplot as plt
```

# Basic Line Plot

```python
# Defining x and y axis
x = np.arange (1,10,0.5)
y = 2*x + 5

# defining the graph 

plt.plot(x,y)
plt.show()
```

# customizing line plot

```python
# Defining x and y axis
x = np.arange (1,10,0.5)
y = 2*x + 5

# changing figure size

fig = plt.figure(figsize = (10,5))

# adding labels

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph')

# plt.plot(x,y)

# plt.show()

# adding legend , grid and customizing line style ....

plt.plot(x,y,linewidth = 3 , linestyle= '-' , color = 'g' , marker = 'o' , alpha = .5)
plt.legend(['line 1'] , loc = 'best')
plt.grid(True)

plt.show()
```

# Subplot

```python
y2 = (y + 5)/2

plt.figure(figsize = (10,5))

plt.subplot(1,3,1) #the figure has 1 row, 2 columns, and this plot is the first plot.
plt.plot(x,y , color = 'b')

plt.title('line 1')


plt.subplot(1,3,2) #the figure has 1 row, 2 columns, and this plot is the second plot.
plt.plot(y,y2 , color = 'g')
plt.title('line 3')

plt.subplot(1,3,3) #the figure has 1 row, 2 columns, and this plot is the second plot.
plt.plot(y2,x , color = 'hotpink')
plt.title('line 3')

plt.show()
```

# Bar Graph

```python
results = {'fardin' : 58 , 'nahid' : 2 , 'shakkhor' : 22 , 'sharif' : 53 , 'chotan' : 61 , 'naeem' : 29}

key = list(results.keys())
value = list(results.values())

plt.bar(key,value)
plt.show()
```

# Bar graph customization

```python
results = {'fardin' : 58 , 'nahid' : 2 , 'shakkhor' : 22 , 'sharif' : 53 , 'chotan' : 61 , 'naeem' : 29}

key = list(results.keys())
value = list(results.values())

plt.xlabel("Name")
plt.ylabel("Roll")
plt.title("Name vs Roll")

plt.bar(key,value , color = 'purple')
plt.show()
```

# Horizontal graph

```python
results = {'fardin' : 58 , 'nahid' : 2 , 'shakkhor' : 22 , 'sharif' : 53 , 'chotan' : 61 , 'naeem' : 29}

key = list(results.keys())
value = list(results.values())

plt.xlabel("Name")
plt.ylabel("Roll")
plt.title("Name vs Roll")

plt.barh(key,value , color = 'black')
plt.show()
```

# Scatterplot 

```python
import matplotlib.pyplot as plt 
%matplotlib inline

a = [10,20,30,40,50,60,70,80]
b = [51,32,23,57,64,15,46,27]
x = [13,22,39,41,30,29,18,22]

plt.scatter(a,b, c = 'r' , s = 200 , edgecolors = 'g' , marker = 'o' , alpha = 0.5)
plt.scatter(a,x, c = 'b' , s = 200 , marker = '4' , alpha = 0.5)

plt.legend(['b' , 'x'] , loc = 'best')
plt.title("Scatter plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)
plt.savefig("Scatterplot.png")

plt.show()
```
